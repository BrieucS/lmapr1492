```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("QpuaQDTCSV1bdImr") as m:

    # Structure for material id
    Sr2SnHg = m.get_structure_by_material_id("mp-867169")
```

```python
#Cr√©ation d'un objet XRDCalculator
xrd = XRDCalculator(1.54060)

#Calcul du mod√®le de diffraction
pattern = xrd.get_pattern(Sr2SnHg)
```

```python
#Angle de diffraction du premier pic
theta1 = pattern.x[0]

#Intensit√© du premier pic
intensit√©1 = pattern.y[0]

#Indices hkl du premier pic
hkl1 = pattern.hkls[0]

#Distance inter-plan
d1_hkl = pattern.d_hkls[0]

print("Premier pic:")
print("Angle : 2ùúÉ = ",theta1,"¬∞")
print("Intensit√© :", intensit√©1)
print("Indices hkl :",hkl1)
print("Distance inter-plan :",d1_hkl)
```

```python
#Angle de diffraction du deuxi√®me pic
theta2 = pattern.x[1]

#Intensit√© du deuxi√®me pic
intensit√©2 = pattern.y[1]

#Indices hkl du deuxi√®me pic
hkl2 = pattern.hkls[1]

#Distance inter-plan
d2_hkl = pattern.d_hkls[1]

print("Deuxi√®me pic:")
print("Angle : 2ùúÉ = ",theta2,"¬∞")
print("Intensit√© :", intensit√©2)
print("Indices hkl :",hkl2)
print("Distance inter-plan :",d2_hkl)
```

```python
#Angle de diffraction du troisi√®me pic
theta3 = pattern.x[2]

#Intensit√© du troisi√®me pic
intensit√©3 = pattern.y[2]

#Indices hkl du troisi√®me pic
hkl3 = pattern.hkls[2]

#Distance inter-plan
d3_hkl = pattern.d_hkls[2]

print("Troisi√®me pic:")
print("Angle : 2ùúÉ = ",theta3,"¬∞")
print("Intensit√© :", intensit√©3)
print("Indices hkl :",hkl3)
print("Distance inter-plan :",d3_hkl)
```
